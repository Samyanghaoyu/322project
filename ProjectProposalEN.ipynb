# CPSC 322 Final Project Proposal  
## 基于 NASA 陨石数据集的陨石降落类型分类研究  
### 组员：Haoyu Yang & Junyou Guo

---

## 1. 项目简介（Project Overview）

本项目使用 NASA 公共开放数据中的 Meteorite Landings 数据集，对陨石在地球上被发现的方式进行分类预测。陨石根据被记录方式分为两类：

- **Fell**：被目击到坠落的陨石  
- **Found**：没有被目击到坠落，而是在地面被发现的陨石  

本项目将通过数据清洗、探索性数据分析（EDA）、特征工程，以及三种分类模型（包含自行实现的随机森林）来预测陨石属于 Fell 或 Found。最终我们将比较三种模型的性能，并报告最优分类器及相关结论。

---

## 2. 数据集描述（Dataset Description）

数据来源：NASA Open Data Portal  
链接：https://data.nasa.gov/Space-Science/Meteorite-Landings/gh4g-9sfh  
格式：CSV  
总记录数：34,513 条  
属性数量：8+  

数据集中包含以下主要字段：

- **name**：陨石名称  
- **id**：陨石编号  
- **recclass**：陨石分类类型（超过 100 类）  
- **mass (g)**：陨石质量（克）  
- **year**：降落或被发现年份  
- **fall**：Fell / Found（本项目的分类目标）  
- **reclat, reclong**：陨石发现的地理坐标  
- **geolocation**：标准化坐标格式  

该数据集规模大、特征丰富，且包含连续与类别属性，适合作为分类任务的数据来源。

---

## 3. 分类任务（Classification Task）

本项目选择的分类任务为：

### 🎯 预测陨石是 **Fell** 还是 **Found**（二分类）

选择该任务的原因：

1. Label 清晰、任务明确，适合课程要求的分类场景。  
2. 数据中包含大量连续属性（mass/year/lat/lon），关系具有复杂性。  
3. 样本分布不平衡（Found 较多），有助于展示欠采样/评估指标等技巧。  
4. 存在噪声（质量极端值、年份缺失等），适合展示数据清洗能力。  
5. Feature 对 label 的关系具有科学意义，例如：  
   - 质量大的陨石更容易被目击  
   - 年代较久远的事件更可能属于 Found  

因此，本项目的分类任务既具有科学意义，也具有足够技术深度。

## 4. 技术内容与实现计划（Implementation & Technical Merit）

### （1）数据预处理

- 处理缺失值（如缺失年份、无效坐标）  
- 清洗异常值（尤其是质量 mass 的极端 outliers）  
- 对 recclass 进行类别分组或编码  
- 将连续属性进行必要的离散化（供 Naive Bayes 使用）  
- 检查地理坐标有效范围  

---

### （2）探索性数据分析（EDA）

我们将生成并解释以下可视化图表：

- Fell vs Found 的类别分布图  
- mass（对数尺度）直方图  
- recclass 的类别分布情况  
- year 与 fall 的关系图  
- 地理经纬度散点图  
- 各特征的相关性热力图（correlation heatmap）

EDA 将帮助我们了解数据内部结构，指导后续的特征工程与模型设计。

---

### （3）分类算法设计与比较

项目将实现并比较以下三种分类器：

1. **Decision Tree（基于 entropy 的 ID3/C4.5 风格）**  
2. **k-Nearest Neighbors 或 Naive Bayes（二选一）**  
3. **MyRandomForestClassifier（自行实现，课程要求）**

随机森林的实现将遵循课程规范：

- 使用 bootstrapping 构建训练/验证集  
- 每个节点随机挑选 F 个可用属性（feature randomness）  
- 生成 N 棵树后，选择验证集表现最好的 M 棵构成最终模型  
- 预测时使用简单多数投票法  

我们将对 N、M、F 进行参数搜索，并多次运行不同随机种子以减少偏差。

## 5. 预期挑战（Anticipated Challenges）

- Fell 与 Found 的类别不平衡问题  
- recclass 类别数量极多，需要特征化处理  
- mass 的数值范围跨度大，需要 log-transform  
- 年份字段存在缺失与格式不标准问题  
- MyRandomForestClassifier 的 TDD 实现有一定复杂度
