## 1. Project Overview

This project uses the NASA â€œMeteorite Landingsâ€ dataset to predict how meteorites were discovered on Earth. Meteorites are recorded in two ways:

- **Found**: The meteorite was discovered on the ground with no witnessed fall.

## 4. Technical Content and Implementation Plan

### (1) Data Preprocessing

- Handle missing values (e.g., missing year or invalid coordinates)

### (2) Exploratory Data Analysis (EDA)

We will generate the following visualizations:

- Class distribution of Fell vs Found  
- Mass distribution (log scale)  
- Recclass category frequency  
- Relationship between year and fall type  
- Latitude/longitude scatterplot

These insights will guide feature engineering decisions.

---

### (3) Classifier Design and Comparison

We will implement and evaluate three classifiers:

1. **Decision Tree (entropy-based)**  
2. **k-Nearest Neighbors or Naive Bayes**  
3. **MyRandomForestClassifier (self-implemented)**  

The random forest will follow the required structure:

- Use bootstrapping to generate train/validation splits  
- At each node, randomly select F candidate attributes  
- Train N trees and select the best M based on validation accuracy  
- Use majority voting for prediction  

We will experiment with multiple combinations of (N, M, F) and repeat runs with different random seeds.
