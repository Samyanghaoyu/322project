Meteorite landings

Data from data.gov, by NASA, in csv format, it include: name, id, nametype, recclass, mass (g), fall, year, reclat, reclong, GeoLocation. 
https://catalog.data.gov/dataset/meteorite-landings

We will try to use recclass, mass, year, reclat, reclong, GeoLocation as information to predict what mass range meteorites are more likely landing and where it more likely landing. Also we can know what materials of meteorites are more likely appear. Problems encountered during the data processing, such as GeoLocation being a string in a special format, make it difficult to extract the numerical information from it. The results of this data analysis will help astronomy enthusiasts find the traces of meteorites more easily, and at the same time, it will assist the government in addressing the potential harm caused by meteorite falls more quickly.
